```python
import numpy as np
from scipy.spatial.distance import pdist, squareform
```

```python
vectors = np.load('minilm_mean_vectors.npz')['vectors'][:10000]
```

```python
pairwise_distances = pdist(vectors, metric='euclidean')

np.save('pairwise_distances.npy', pairwise_distances)
```

```python
pairwise_distances = np.load('pairwise_distances.npy')
```

```python
import matplotlib.pyplot as plt

hist, bin_edges = np.histogram(pairwise_distances, bins=30)

plt.hist(pairwise_distances, bins=50, edgecolor='black', color='purple')
plt.title('Pairwise Euclidean (L2) Distances Between Vectors')
plt.xlabel('Vector Distance')
plt.ylabel('Frequency')
plt.show()

print(f"Average distance: { np.mean(pairwise_distances) }")
```
